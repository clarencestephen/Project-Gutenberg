```python
# imports
from glob import glob
from pathlib import Path
import os
```

```python
twos = glob('texts/* 2.txt')
dashes = glob('texts/*-*.txt')
texts = glob('texts/*.txt')
```

```python
clean_twos = [item.replace('texts/','').replace('.txt','').replace('-8','').replace('-0','').replace(' 2','') for item in twos]
len(clean_twos)
```

```python
clean_dashes = [item.replace('texts/','').replace('.txt','').replace('-8','').replace('-0','').replace(' 2','') for item in dashes]
len(clean_dashes)
```

```python
clean_texts = [item.replace('texts/','').replace('.txt','').replace('-8','').replace('-0','').replace(' 2','') for item in texts]
unique_texts = set(clean_texts)
len(unique_texts)
```

```python
len(texts)-len(dashes)-len(twos)
```

```python
file_paths = []
for name in unique_texts:
    file_paths.append("texts/"+name+"*.txt")

file_paths
```

```python
import shutil

for file in unique_texts:
    f = glob('texts/'+file+'.txt') + glob('texts/'+file+' *.txt') + glob('texts/'+file+'-*.txt')
    f = f[0]
    os.rename(f, "texts/"+f) #move to subdirectory
```
